```julia
f = x-> 1.5 - exp(-0.25*(x[1]-3)^2)  - 0.75*exp(-0.5*(x[1]-7)^2)
∇ = x-> [0.75*exp(-0.5*(x[1]-7)^2)*(x[1]-7) + exp(-0.25*(x[1]-3)^2)*0.5*(x[1]-3)]

a = 0.0
b = 7.0

axis_xlo = 0
axis_xhi = 7.5
ymin = 0.45
ymax = 1.65

x = [0.0]
d = [1.0]
β = 0.5
σ = 0.9
```

```julia
plot(a:0.1:b, f, xlims = (axis_xlo, axis_xhi), 
    yticks = false, xticks = false, 
    label = "f(x)", framestyle = :box, ylabel = "y", xlabel = "α")
plot!([a,1.5], α->f(x) + α*∇(x)'*d, label = "β = 1.0") 
plot!([a,b], α->f(x) + β*α*∇(x)'*d, label = "β = 0.5")
plot!([a,b], α->f(x) + 0*α*∇(x)'*d, label = "β = 0.0") 
```
